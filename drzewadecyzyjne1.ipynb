```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

# Drzewa decyzyjne

### Ćwiczenie 1: dane 2 księżyce

```python
moons = make_moons(n_samples = 150, noise = .1, random_state = 17)
```

```python
X = moons[0]
y = moons[1]
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], marker = "o", c = "red", edgecolor = "black", s = 50)
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker = "v", c = "blue", edgecolor = "black", s = 50)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 17)
```

```python
model1 = DecisionTreeClassifier(random_state = 17)
model1.fit(X_train, y_train)
```

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
fig = plt.figure(figsize = (15, 9))
tree.plot_tree(model1, feature_names = ["X1", "X2"], class_names = ["0", "1"], filled = True)
plt.show()
```

```python
model1.score(X_train, y_train)
```

```python
model1.score(X_test, y_test)
```

```python
model2 = DecisionTreeClassifier(random_state = 17, max_depth = 3).fit(X_train, y_train)
```

```python
model2.score(X_train, y_train)
```

```python
model2.score(X_test, y_test)
```

# Lasy losowe

```python
model3 = RandomForestClassifier(n_estimators = 10, max_features = 1, max_depth = 4, random_state = 17)
model3.fit(X_train, y_train)
```

```python
model3.score(X_train, y_train)
```

```python
model3.score(X_test, y_test)
```

```python
pred3 = model3.predict(X_test)
```

```python
confusion_matrix(y_test, pred3)
```

```python
model4 = RandomForestClassifier(n_estimators = 100, random_state = 17)
model4.fit(X_train, y_train)
```

```python
model4.score(X_test, y_test)
```
